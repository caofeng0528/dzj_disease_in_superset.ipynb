```python
con = get_con_db()
sql = '''
    select disease,apply_tag,old_remain from dw.golden_standard_superset
    where apply_tag = 1 and old_remain is null
'''
superset = pd.read_sql(sql,con)
superset['disease'] = superset['disease'].apply(lambda x: x.strip())
superset
```

```python
disease = pd.read_excel('./upload_data/常见疾病列表v1.0(1).xlsx')
disease_row_null = disease.dropna(axis=0, how='all')
disease_col_null= disease_row_null.dropna(axis=1, how='all')
data = disease_col_null.drop(disease_col_null.tail(2).index,inplace=True)
data1 = disease_col_null.drop('Unnamed: 29',axis = 1,inplace = True) #axis参数默认为0
disease_col_null['disease_in_superset'] = ''
disease_col_null['疾病'] = disease_col_null['疾病'].apply(lambda x: x.strip())
disease_col_null
```

```python
for i,row in disease_col_null.iterrows():
    if row['疾病'] in list(superset['disease']):
        disease_col_null.loc[disease_col_null['疾病'] == row['疾病'],'disease_in_superset'] = 1
    else:
        continue
disease_col_null
```
